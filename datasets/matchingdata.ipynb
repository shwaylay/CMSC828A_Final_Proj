```python
import pandas as pd
```

```python
df_cur  = pd.read_csv('layoffs_and_stock_info_CLEANED.csv')
df_new = pd.read_excel('tech_layoffs.xlsx')
```

```python
df_cur
```

```python
df_new
```

```python
company_list_1 = df_cur['company'].str.lower().replace(" ", "")
unique_list_1 = list(dict.fromkeys(company_list_1))
 
```

```python
company_list_2 = df_new['Company'].str.lower().replace(" ", "")
unique_list_2 = list(dict.fromkeys(company_list_2))
```

```python
unique_companies = list(set(unique_list_1) & set(unique_list_2))
```

```python
unique_companies
company_symbols = [df_cur[df_cur['company'].str.lower().replace(" ", "") == company]['Symbol'].values[0] for company in unique_companies]
```

```python
company_to_symbol = dict(zip(unique_companies, company_symbols))
```

```python
df_new = df_new[df_new['Company'].str.lower().replace(" ", "").isin(unique_companies)]
```

```python
df_new["Symbol"] = [company_to_symbol[company] for company in df_new['Company'].str.lower().replace(" ", "")]
```

```python
df_new["Date_datetime"] = pd.to_datetime(df_new['Date_layoffs'])
```

```python
# Sort data by date_datetime
df_new = df_new.sort_values(by='Date_datetime')
```

```python
df_new
```

```python
df_new_nona=  df_new.dropna(subset=['Company_Size_before_Layoffs', "Percentage"])
```

```python
num_companies = set(df_new_nona["Symbol"])
```
