```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stock = pd.read_csv("amazon_stock.csv", parse_dates=['Date'])
layoff = pd.read_csv("amazon_layoff.csv", parse_dates=["date"])
```

```python
fig, ax1 = plt.subplots()
ax1.plot(stock['Date'], stock['Open'])
ax2 = ax1.twinx()
ax2.plot(layoff['date'], layoff['total_laid_off'], color="red")
plt.show()
```
